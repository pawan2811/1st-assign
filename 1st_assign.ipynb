````python
#6  How do conditional statements work in Python? Illustrate with examples
'''Conditional statements in Python allow you to execute certain pieces of code based on whether a condition is true or false. The primary conditional statements in Python are `if`, `elif`, and `else`.



1. if` statement: This is used to test a specific condition. If the condition is true, the code block under the `if` statement is executed.
2. elif` statement: This stands for "else if" and allows you to check multiple conditions. If the preceding `if` condition is false, Python checks the `elif` condition. You can have multiple `elif` statements.

3. else` statement: This is used to execute a block of code when none of the preceding conditions are true. It comes at the end of the `if` and `elif` blocks.
'''
### Example 1: Basic `if` statement


x = 10
if x > 5:
    print("x is greater than 5")


### Example 2: `if-elif-else` statement

x = 10
if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but less than or equal to 15")
else:
    print("x is 5 or less")


### Example 3: Nested `if` statements


x = 10
y = 20

if x > 5:
    if y > 15:
        print("x is greater than 5 and y is greater than 15")
    else:
        print("x is greater than 5 but y is not greater than 15")
else:
    print("x is not greater than 5")


### Example 4: Multiple `elif` statements

score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")


#In these examples, the conditions within the `if`, `elif`, and `else` statements are evaluated in order. Once a true condition is found, the corresponding block of code is executed, and the rest are skipped. If none of the conditions are true, the `else` block is executed.
````

````python
#7Describe the different types of loops in Python and their use cases with examples
'''
In Python, there are two primary types of loops: `for` loops and `while` loops. Each has distinct use cases and functionalities.

### 1. `for` Loop

The `for` loop is used to iterate over a sequence (such as a list, tuple, dictionary, set, or string) or any other iterable object. It allows you to execute a block of code multiple times.

#### Use Cases
- Iterating through a list or tuple.
- Accessing keys and values in a dictionary.
- Repeating a block of code a specific number of times.
'''
#Example 1: Iterating through a list


fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    print(fruit)


# Example 2: Iterating through a string


word = "Python"

for letter in word:
    print(letter)

# Example 3: Using `range()` to iterate a specific number of times


for i in range(5):
    print(i)

# Example 4: Iterating through a dictionary


student_scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78}

for name, score in student_scores.items():
    print(f"{name}: {score}")


### 2. `while` Loop

#The `while` loop is used to repeat a block of code as long as a specified condition is true. It is useful for situations where the number of iterations is not known beforehand.

# Repeating a block of code until a certain condition is met.
#Implementing retry logic.
# Reading data until the end of a file.

#### Example 1: Basic `while` loop

count = 0

while count < 5:
    print(count)
    count += 1


#### Example 2: `while` loop with a condition

number = 10

while number > 0:
    print(number)
    number -= 1


#### Example 3: Using `break` in a `while` loop

```python
count = 0

while True:
    print(count)
    count += 1
    if count == 5:
        break

#### Example 4: Using `continue` in a `while` loop

count = 0

while count < 10:
    count += 1
    if count % 2 == 0:
        continue
    print(count)
'''Summary

- **`for` Loop**: Best for iterating over a known sequence of elements or a specific number of times.
- **`while` Loop**: Best for repeating a block of code as long as a certain condition is true, where the number of iterations is not known beforehand.

Each type of loop has its strengths, and choosing the right one depends on the specific use case and the nature of the task you are trying to accomplish.'''
````
